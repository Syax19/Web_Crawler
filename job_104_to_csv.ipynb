```python
import glob

job_data = []
for path in glob.glob("./job_104/*.txt"):
    with open(path, "r", encoding="utf-8") as f:
        try:
            job_data.append(f.read().split("\n"))
        except:
            print("Some error happend...")
            
print(job_data)
```

```python
import pandas as pd

columns = ["Job_Company", "Job_Title", "Job_Tenure", "Education", "Job_Content", "Salary", "Job_Url", "Company_Address", "Company_Url"]

df = pd.DataFrame(data=job_data, columns=columns)
df
```

```python
def columnfilter(x):
    str_a = "："
    if str_a in x:
        return x.split("：")[1]
    else:
        return x
```

```python
df["Company_Address"] = df["Company_Address"].apply(columnfilter)
```

```python
df.to_csv("./job_104_table.csv", index=False, encoding="utf-8-sig")
```
